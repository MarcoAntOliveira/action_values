```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def obtem(ticker, start, end):
    data = yf.download(ticker, start="2023-01-01", end="2023-11-18")
    return data
```

```python
def graph(data, labelx, labely):
    plt.figure(figsize=(10, 6))
    plt.plot(data[labelx], data[labely])
    plt.title(f"{labelx}")
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.grid(True)
    plt.show()
```
